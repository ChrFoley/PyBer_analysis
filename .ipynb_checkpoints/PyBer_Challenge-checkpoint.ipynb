```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
# total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  total fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
# calculate the average fare per ride
average_fare_ride = total_fares / total_rides
average_fare_ride.head()
```

```python
# calculate the average per driver
average_per_driver = total_fares / total_drivers
average_per_driver.head()
```

```python
# create the summary dataframe
summary_df = pd.DataFrame({
    "Total rides": total_rides,
    "Total drivers": total_drivers,
    "Total fares": total_fares,
    "Average fare per ride": average_fare_ride,
    "Average fare per driver": average_per_driver})

summary_df.head()
```

```python
# format the dataframe
summary_df["Total fares"] = summary_df["Total fares"].map("${:,.2f}".format)
summary_df["Average fare per ride"] = summary_df["Average fare per ride"].map("${:.2f}".format)
summary_df["Average fare per driver"] = summary_df["Average fare per driver"].map("${:.2f}".format)

summary_df.head()
```

```python
#delete the index
summary_df.index.name = None
summary_df
```

Building the chart

```python
#  Rename columns 
pyber_data_df.rename(columns={'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'}, inplace=True)
pyber_data_df.head()
```

```python
# set index to the "date" column.
pyber_data_df.set_index(pyber_data_df['Date'], inplace=True)
pyber_data_df.head()
```

```python
# Referenced from solution. 


#  adjust df for fares and include only the 'Date', 'City Type', and'Fare' columns.
pyber_cities_fares = pyber_data_df[['Date','City Type' ,'Fare']].copy()

# delete the  'date' column. 
pyber_cities_fares.drop(['Date'], axis=1, inplace=True)

# Set the index  
pyber_cities_fares.index = pd.to_datetime(pyber_data_df.index)
pyber_cities_fares.head(10)
```

```python
# Calculate the sum() of fares by the type of city and date using groupby() to create a Series. 
sum_fare_type = pyber_cities_fares.groupby(["City Type", "Date"]).sum()["Fare"]
sum_fare_type.head(10)
```

```python
# create the DataFrame
sum_fare_type = pd.DataFrame(sum_fare_type)
sum_fare_type.head(10)
```

```python
# Reset the index to create the pivot table
sum_fare_type = sum_fare_type.reset_index()
sum_fare_type.head(10)
```

```python
# Create the pivot table with date as index and city type as columns 
sum_fare_type_pivot = sum_fare_type.pivot(index="Date", columns="City Type") ["Fare"]
sum_fare_type_pivot.head()
```

```python
# Referenced from the solution
# create a new data frame with the dates provided using the loc function
fares_Jan_April = sum_fare_type_pivot.loc['2019-01-01':'2019-04-28']
fares_Jan_April.head(10)
```

```python
# referenced from the solution.
# create a new dataframe with the previous by setting weekly bins and calculate the sum
weekly_fares_df = fares_Jan_April.resample('W').sum()
weekly_fares_df.head(10)
```

```python
# import the stlye from matplotlib
from matplotlib import style
style.use('fivethirtyeight')
```

```python
# create the line chart, usng the fivethirtyeight style
ax = weekly_fares_df.plot(figsize = (30, 10))
ax.set_title("Total Fares by City Type")
ax.set_xlabel("Month")
ax.set_ylabel("Fare $USD")
plt.savefig("Analysis/PyBer_Challenge_fare_summary.png")
plt.show()
```
